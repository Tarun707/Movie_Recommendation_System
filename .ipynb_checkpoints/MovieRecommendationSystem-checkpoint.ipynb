```python
# import ast
# data_1['genres'] = data_1['genres'].map(lambda x: ast.literal_eval(x))
# data_1['cast'] = data_1['cast'].map(lambda x: ast.literal_eval(x))
# data_1['crew'] = data_1['crew'].map(lambda x: ast.literal_eval(x))


from ast import literal_eval

# Converting object data type to a list.
# literal_eval has a problem with null values so I replaced null with None.

# data_1['cast'] = data_1['cast'].apply(literal_eval)
# data_1['genres'] = data_1['genres'].apply(literal_eval)
# data_1['crew'] = data_1['crew'].apply(literal_eval)

data_1['cast'] = data_1['cast'].str.replace('null', 'None')
data_1['cast'] = data_1.cast.apply(literal_eval)
data_1['genres'] = data_1['genres'].str.replace('null', 'None')
data_1['genres'] = data_1.genres.apply(literal_eval)
data_1['crew'] = data_1['crew'].str.replace('null', 'None')
data_1['crew'] = data_1.crew.apply(literal_eval)
```

```python
def make_genres(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.nan
    else:
        return (st.join(gen))
```

```python
data_1['genres_list'] = data_1['genres'].map(lambda x: make_genres(x))
```

```python
data_1['genres_list']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.nan
    else:
        return (casts[0])
```

```python
data_1['actor_1_name'] = data_1['cast'].map(lambda x: get_actor1(x))
data_1['actor_1_name']
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.nan
    else:
        return (casts[1])
```

```python
data_1['actor_2_name'] = data_1['cast'].map(lambda x: get_actor2(x))
data_1['actor_2_name']
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data_1['actor_3_name'] = data_1['cast'].map(lambda x: get_actor3(x))
data_1['actor_3_name']
```

```python
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data_1['director_name'] = data_1['crew'].map(lambda x: get_directors(x))
```

```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn import naive_bayes
from sklearn.metrics import roc_auc_score,accuracy_score
import pickle
```

```python
nltk.download('stopwords')
```

```python
reviews_data = pd.read_csv('reviews.txt',sep = '\t', names =['Reviews','Comments'])
```

```python
reviews_data
```

```python
stopset = stopwords.words('english')
```

```python
tfidf_vectorizer = TfidfVectorizer(use_idf = True,lowercase = True, strip_accents='ascii',stop_words=stopset)
```

```python
X = tfidf_vectorizer.fit_transform(reviews_data['Comments'])
```

```python
print(X)
```

```python
y = reviews_data.Reviews
```

```python
y
```

```python
with open('transform.pkl', 'wb') as file:
    pickle.dump(tfidf_vectorizer, file)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
clf = naive_bayes.MultinomialNB()
clf.fit(X_train,y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.score(X_train, y_train)
```

```python
with open('nlp_model.pkl', 'wb') as file:
    pickle.dump(clf, file)
```
